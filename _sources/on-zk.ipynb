## 计算 computation

我们能计算啥，不能计算啥，什么方便计算，什么难于计算？

这个概念的清晰化，至少很大程度归功于图灵。

### 图灵机模型

模型由如下部件组成：

1. TAPE: 一个纸带， **无限长**，然后被分成了一格一格的，格子里面可以写上一个字母
>
> 无限长表明这玩意儿还是挺理想化的，但是因为别的部件都是有限的，所以TAPE实际被用到的地方是有限的
>
> 所谓的字母，是一种抽象意义的字母表，不一定是ABCD...，最简单的字母表的例子是： 1) 只有一个字母； 2) 只有两个字母，比如 0 和 1
>
> 格子上的内容，有些是作为输入值最开始就写好的，有些（比如很远很远的地方）可以认为是某个固定的值，比如是0
>
> TAPE应该就没有别的细节了，如果你在纠结别的细节，说明你想多了。比如格子是不是等宽的，这个毫无意义 :)
>

2. HEAD: 读写头，任何时刻都处于TAPE的某个格子上，因此可以读取格子的内容，也可以修改格子的内容

3. STATE: 一个记录当前状态的寄存器
>
> 图灵机的状态个数是有限的，因此STATE是可以方便的物理实现
>
> STATE可以简单的理解为是比如 1,2,3...，但为了方便表示，也可以是 INIT,RUNNING,FINISH 之类的有意义的东西，当然无论是什么，最终都可以用个整数来表示

4. TABLE: 控制规则表。他根据前面几个部件的状态，然后决定下一个动作。下一步动作按顺序是: 1) 不写入，或者写入一个字母; 2) 保持HEAD不动，或者向左向右移动； 3) STATE保持不变，或者变为另外一个状态
>
> 所以在这个模型中，时间是离散的，或者类似计算机中一样，是一个时钟周期一个时钟周期往前推进的，一个tick一个tick的往前走的。每个tick，图灵机按照规则表，读取数据，并产生然后执行相应的动作
>
> TABLE可以认为是一行一行的，每行前面几列是： HEAD下面的格子的内容，STATE的内容，这些是规则的输入； 后面几列是上述的3个动作，这些动作按顺序完成。
>

于是图灵宣称，我们能计算的就是图灵机能计算的，图灵机能计算的就是我们能计算的。这玩意儿，先不管说是不是正确，但至少非常准确的刻画了一类可以计算的东西。

至于这个东西是否准确的表达了我们人类的计算概念，只能说，这是一个哲学问题。当然时间能够考验一切，无数聪明大脑的批判能够考验一切。历史证明，图灵模型是经得起考验的。**In time, in history, in rational criticism, in bet-on, we trust.**
>
> 容易看出，图灵机能够计算的，我们人类应该是毫无压力可以计算的，所以图灵机模型是 小于等于 我们人类的计算概念的。

> 那反过来，到底有没有那种，人类原则上能够进行计算的，但是图灵机却不能计算的东西呢？或者说，在我们这个宇宙，物理意义上能够进行计算的，但是图灵机却算不出来的呢？还真有，至少逻辑上是有的，比如物理上可以任意精度下进行实数运算，图灵机不行。例如，一个直径为1的圆的周长是 $\pi$ ，那么把直径放大一倍，周长自然就是 $\pi + \pi$，这在物理上毫无压力，或者说你拿个尺子去量毫无压力（精度是你能力问题，不是物理问题），但是用图灵机，你连完整表达一个无理数的办法都没有。。。当然，这种例子，完全不影响人类认可图灵机模型，对吧？

## 信息 information

何为信息？

我告诉我女儿是： 一加一等于二。我传达信息给她了吗？ 分别考虑我女儿 1岁 和 3岁 和 10岁 的情形呢？

我女儿调皮的时候，我想揍她，她问我为什么，我说： 我是你爸爸。 请问我传达信息给她了吗？传达了多少信息？

如同上述 **计算** 一样，**信息** 这种人类直觉的东西，如何准确定义？

直到 **香农** 横空出世，把信息 **量化** 为 （简化版）：

1. 我们说的 **信息** 是和某个 **事件** 的发生概率相关。
>
> 比如明天会不会下雨，我们的预判是 **一半概率下雨，一半概率不下雨**
>

2. **信息源** 把一个 **消息** 通过 **信道** 传递给 **我们** 。
>
> 比如信息源是天气预报，信道是广播，消息是 **明天下雨**
>

3. 此 **消息** 所传达的信息量是该事件的 **不确定性** 的消除的量。用数学用语就是，**熵** 的减少。 
>
> 比如，在我们预判中，下不下雨的概率都是一半，对应的 **熵** 是 $ H(X) = \frac{1}{2} * \log_2{\frac{1}{1/2}} + \frac{1}{2} * \log_2{\frac{1}{1/2}} = 1$ 。经过天气预报一提醒，我们知道一定会下雨，对应的 **熵** 是 $ H(X') = 1 * \log_2{\frac{1}{1}} = 0$ 。熵的较少就是 $1 - 0 = 1$ ，这就是该消息所传达的信息量。
>
> 换个例子，如果天气预报说的是，明天下雨概率 $1/3$ ，不下雨概率是 $2/3$，那么该消息传达的信息量是 $1 - H(X') = 1 - (\frac{1}{3} * \log_2{\frac{1}{1/3}} + \frac{2}{3} * \log_2{\frac{1}{2/3}}) \approx 0.0817$
>
> 因此，容易看出，第二种情况下，该消息传递的信息量更少了。还容易看出，如果我们改变初始条件，消息的传递是可能导致 **熵** 增加的，也就是说，该消息传递了一个 **负的信息量**。

看起来，该系统和通信系统息息相关，没错，该理论最开始是在贝尔实验室搞出来的，而且论文题目是 《A Mathematical Theory of Communication》 <https://en.wikipedia.org/wiki/A_Mathematical_Theory_of_Communication> 。

香农何德何能，有资格下这样一个定义，还让所有追随者趋之若鹜？难道这就是信息的完整定义？难道这就囊括了我们所期待准确定义的那些 **信息** ？我想是应该没有，比如我前面举的几个例子，就无法合理的对应到香农的这个定义中来。

香农自然不是神，香农在构建其理论大厦的时候，所需要的大部分砖块儿其实已经唾手可得，但是香农的思想又是引领时代的，真因为香农的思想让 **信息** 的定义清晰化了，所有的**砖块**就只有考古学的意义了。

香农是幸运的，他的这临门一击，正切中那个时代的通信的迫切需求，他的理论实际上并没有碰到明显的阻碍，自然就被人们所接受了。

但是，批判的来看，难道就只有香农所定义的信息才是信息？难道我之前列举出的生活种种就没有信息？难道有了香农，有了香农的信息论，人类自古以来通过声音，通过语气，通过别的途径，所传达的那些无法准确量化的信息就不是信息了？不应该是这样，只是香农声势浩大，夺下了 **信息** 在数学上的那块阵地。只是，在包括通信，包括计算机的某些领域中，我们承认，香农的理论目前是靠谱的，是值得的，我们批判的接受这一点。我们依然在下一个赌注。

# 知识 和 零知识

**知识** 同样难以量化，不过幸运的是，**香农** 帮我们承担了一切。

如文首所述，如果我们开心的承认香农的 **信息论**，我们大起胆子，不再纠结，我们就很容易给知识下一个定义。我们把批评留个香农，我们轻装上阵。

类似 **信息** 的定义，我们大胆的来定义知识。

1. 知识是关于某个实体的，某个事物的
>
> 比如，我想知道你有多少钱，这就是关于某个数的知识。比如我想知道，明天早上8点，微软的股价是多少，这也是关于某个数的。比如我想知道特朗普和普金的某次通话的内容，这就是关于某段语言的知识。
>

2. 我们用概率分布来度量知识
>
> 比如，你扔了一颗骰子，然后捂住。这个时候，对于这颗骰子上的点数，除了知道是1-6之外，一无所知。所以我对这颗骰子上的点数的知识，可以准确的表达为概率分布 $\{\frac{1}{6}, \frac{1}{6}, \frac{1}{6}, \frac{1}{6}, \frac{1}{6}, \frac{1}{6}\}$ 。
>
> 然后，你告诉我，骰子的点数是奇数，于是我的知识更新为 $\{\frac{1}{3}, 0, \frac{1}{3}, 0, \frac{1}{3}, 0\}$ 。概率分布发生了变化，表示我的知识发生了变化，于是你给我传递了知识。
>
> 但是，如果，你告诉我的是，骰子的点数 小于 10，我的知识并不会发生任何更新，于是你根本没有告诉我新的知识。

3. 知识的传递，导致度量知识的概率分布的改变

## 零知识

零知识更多的是指出一个过程没有发生知识的传递，比如上文中 “你告诉我的是，骰子的点数 小于 10” 的情形。

似乎零知识很简单，对吧？定义很清楚，对吧？那我们来看这个例子：
>
> 你拥有一个比特币地址，上面有100个比特币，然后你把地址告诉了我，让我给你打钱。请问这个过程，你有没有传递 **关于你的私钥的知识** 给我。
>

你会说 **NO**，不然那还了得，但实际上，答案是 **YES**，为什么？

        address your_addr;
        for(bigint p = 0;p < 2**256;p++) {
            pubkey pub = private_key_to_pub_key(p)
            addr = pubkey_to_addr(pub)
            if(addr == your_addr) {
                return p
            }
        }
        return "you cheat"

上面的程序，将在有限的时间内，从你的地址计算出你的私钥。程序计算结束，我将拥有你的完整私钥。所以，答案必然是 **YES** 。

但是，为什么，你不会有丝毫恐慌？原因是什么？是因为这个方法适合于所有的人？你不担心你的私钥我更容易算出来的么？比如，如果你的地址的私钥是 $p < 10000$ ，那我的程序将可以在几微秒就能算出你的私钥，这不是很恐怖的吗？
