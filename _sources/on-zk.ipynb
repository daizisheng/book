## 进步

我曾经天真的以为，数学这种科学，除了公理，一切都是逻辑的推论，那么只要付出时间和汗水，没有什么是学不懂的。但事实似乎并非如此。

上过大学，学过一点高等数学的都知道，牛顿的伟大成就之一就是发明了微积分。牛顿真的是被一个莫名其妙的苹果砸中脑袋，从而神奇的发明了一系列科学和数学原理吗？应该不是，至少，微积分，应该是牛顿站在前人的基础上，厘清混沌，付出勇气，大胆的迈出那一步，然后接受质疑，勇猛往前，从而奠定了自己微积分发明者的历史地位。但即使如此，人类充分搞明白微积分，人类真正的相信自己已经逃脱了那 **无穷小除以无穷小** 的梦魇，从数学大厦中排除了那 **一会儿是0，一会儿非0** 的幽灵，也是几百年之后的事情了。幸之又幸，我们生活在可以简单接受微积分的思想毫无压力的时代，我们在课堂上相信老师的言之凿凿，然后用微积分来计算奇怪形状的长度、面积和体积，用它来解答物理问题并用实验验证，期末考试拿到高分，毕业拿到很好的GPA，我们都有光明的未来。除非，我们进入的是数学系，学的是基础数学，不然我们会认为微积分是理所当然，那些 **梦魇**，那些 **幽灵**， 与我何干。

但，如果我们来看另外一个例子。

19世纪，**康托** 发明了集合论，其中一个结论是证明了 **自然数和正有理数** 一样多。他采用的技巧 **对角线法** 如下：

> 1. 自然数 1,2,3,... 是正有理数的一个子集；
>
> 2. 正有理数如果按照下面表格进行排序，我们将可以把所有正有理数对应到自然数；
>
> 因此，自然数和正有理数一样多
>

$$
\begin{array}{|c|c|c|c|c|c}
\hline \\
1: \frac{1}{1} & 3: \frac{2}{1} & 6: \frac{3}{1} & 10: \frac{4}{1} & 15: \frac{5}{1} & ... \\ \\ \hline \\
2: \frac{1}{1} & 5: \frac{2}{1} & 9: \frac{3}{1} & 14: \frac{4}{1} & ... & ... \\ \\ \hline \\
4: \frac{1}{1} & 8: \frac{2}{1} & 13: \frac{3}{1} & ... & ... & ... \\ \\ \hline \\
7: \frac{1}{1} & 12: \frac{2}{1} & ... & ... & ... & ... \\ \\ \hline \\
11: \frac{1}{1} & ... & ... & ... & ... & ... \\ \\ \hline \\
... & ... & ... & ... & ... & ...
\end{array}
$$

这是什么鬼，如果不是在课堂上学到，我会认为这个人是个疯子，是个精神病。如果我相信上帝，我会认为这个人是在亵渎上帝，要被处以火刑。

事实上，维特根斯坦说：
>
> 集合论是假借通过数学而有害处的方言，是可笑和错误的完全无稽之谈。
>

当时欧洲最杰出的数学家之一克罗内克，也是康托的老师，表示
>
> 无法接受一个“冒充内行的‘牛皮匠’和带坏青年的‘教唆者’”与自己共事。在学生面前，他竭尽全力批评康托尔的工作，声称“康托尔的成果是个骗局”。
>

结果是，
>
> 人们认为康托将数学带到了精神病院，可没想到的是，最后被带到精神病院的，反而是康托
>

最后康托死在了精神病院。

后来，大卫·希尔伯特说：
>
> “没有人能够把我们从康托尔建立的乐园中赶出去。”
>

集合论在现在已经广为被人们所接受，如果要去学习集合论，我们自然有信心类似 **对角线法** 的奇谈怪论习以为常，欣然接受。但康托同时代的人就没这么幸运，他们认为这是魔鬼，这是病，这是将数学带入精神病院的妖孽。

那是什么让曾经如瘟疫一样的奇谈怪论，在今天变成了公认的真理，是逻辑的自然推论吗？不是。是时间，是其有用，是100多年的数学实践，是无数基于它的没有引发矛盾的研究，是无数基于它所产生的新的知识以及新的知识的美妙和有用性。我们是在下一个赌注，我们认为康托的集合论虽然是反直觉的，是无法通过逻辑推导出来的，但是却是可以信任的。我们信任它，基于它构建新的高楼大厦，我们相信高楼大厦将长期屹立不倒。但是，如果我们赌输了，我们的高楼大厦出现了裂痕，我们就会反过来推导出集合论的缺陷，我们就会获得菲尔兹奖，我们会经历新的短暂混乱，但终究会跨越灾难，离真理更进一步。这个赌注，值得。

所以，为什么我们能相信康托，不是因为 **in math, we trust** ，不是因为 **in logic, we trust** ，而是

>
> **In time, in history, in rational criticism, in wager, we trust.**
>

这是我认为的真正自学数学的难点，这是课堂上老师潜移默化传递给你的信心，只有这个信心，加上时间和汗水，你才能真正搞懂数学。

后文讲到的 **计算，信息，零知识** 等一堆概念，或多或少都是类似 **微积分，集合论** 的奇谈怪论，要跨越理解的鸿沟，你需要相信，需要在 **理性批判主义** 的意义下相信。


## 计算 computation

我们能计算啥，不能计算啥，什么方便计算，什么难于计算？

这个概念的清晰化，至少很大程度归功于图灵。

### 图灵机模型

模型由如下部件组成：

1. TAPE: 一个纸带， **无限长**，然后被分成了一格一格的，格子里面可以写上一个字母
>
> 无限长表明这玩意儿还是挺理想化的，但是因为别的部件都是有限的，所以TAPE实际被用到的地方是有限的
>
> 所谓的字母，是一种抽象意义的字母表，不一定是ABCD...，最简单的字母表的例子是： 1) 只有一个字母； 2) 只有两个字母，比如 0 和 1
>
> 格子上的内容，有些是作为输入值最开始就写好的，有些（比如很远很远的地方）可以认为是某个固定的值，比如是0
>
> TAPE应该就没有别的细节了，如果你在纠结别的细节，说明你想多了。比如格子是不是等宽的，这个毫无意义 :)
>

2. HEAD: 读写头，任何时刻都处于TAPE的某个格子上，因此可以读取格子的内容，也可以修改格子的内容

3. STATE: 一个记录当前状态的寄存器
>
> 图灵机的状态个数是有限的，因此STATE是可以方便的物理实现
>
> STATE可以简单的理解为是比如 1,2,3...，但为了方便表示，也可以是 INIT,RUNNING,FINISH 之类的有意义的东西，当然无论是什么，最终都可以用个整数来表示

4. TABLE: 控制规则表。他根据前面几个部件的状态，然后决定下一个动作。下一步动作按顺序是: 1) 不写入，或者写入一个字母; 2) 保持HEAD不动，或者向左向右移动； 3) STATE保持不变，或者变为另外一个状态
>
> 所以在这个模型中，时间是离散的，或者类似计算机中一样，是一个时钟周期一个时钟周期往前推进的，一个tick一个tick的往前走的。每个tick，图灵机按照规则表，读取数据，并产生然后执行相应的动作
>
> TABLE可以认为是一行一行的，每行前面几列是： HEAD下面的格子的内容，STATE的内容，这些是规则的输入； 后面几列是上述的3个动作，这些动作按顺序完成。
>

于是图灵宣称，我们能计算的就是图灵机能计算的，图灵机能计算的就是我们能计算的。这玩意儿，先不管说是不是正确，但至少非常准确的刻画了一类可以计算的东西。

至于这个东西是否准确的表达了我们人类的计算概念，只能说，这是一个哲学问题。当然时间能够考验一切，无数聪明大脑的批判能够考验一切。历史证明，图灵模型是经得起考验的。In time, in rational criticism, in history, we trust.
>
> 容易看出，图灵机能够计算的，我们人类应该是毫无压力可以计算的，所以图灵机模型是 小于等于 我们人类的计算概念的。
>
> 那反过来，到底有没有那种，人类原则上能够进行计算的，但是图灵机却不能计算的东西呢？或者说，在我们这个宇宙，物理意义上能够进行计算的，但是图灵机却算不出来的呢？还真有，至少逻辑上是有的，比如物理上可以任意精度下进行实数运算，图灵机不行。例如，一个直径为1的圆的周长是 $\pi$ ，那么把直径放大一倍，周长自然就是 $\pi + \pi$，这在物理上毫无压力，或者说你拿个尺子去量毫无压力（精度是你能力问题，不是物理问题），但是用图灵机，你连完整表达一个无理数的办法都没有。。。当然，这种例子，完全不影响人类认可图灵机模型，对吧？
>

### 图灵机万能吗？

不可能，世界不会简单到 **真理=可计算** ， 比如估计谁也算不出明天的股票价格。。。 当然，这里不说这个。我们得来一个，能精准描述，但是图灵机就是算不出来的东西。

最著名的例子莫过于是所谓的 **停机问题** ，他可以这样描述：

请你设计一个图灵机，他的名字叫 **U** ，就是 **universal** 的 **U** ，可以叫做万能图灵机，他需要干啥呢。我们把任何用户给的图灵机 **I**，就是 **input** 的 **I** ，编码好，然后写在你的 **TAPE** 上，然后你的 **U** 要干的事儿，就是计算出，我这个 **I** 到底能不能停下来，或者准确一点说，我的 **I** 的 **STATE** 能不能在某个时刻进入到某个指定的值上，比如 **FINISH** 。

这说起来，好像有点拗口，我们直接用现在的计算机程序举个例子：

你能不能写一个比如 java 程序，然后我给你输入一个 c 程序，你告诉我，我这个 c 程序会不会在有限的时间内完成。当然你的程序也得在有限时间完成。

在具体一点，比如我的程序如下的话，你就输出 **YES**

    int main() { return 0; }

但如果如下的话，你就输出 **NO**

    int main() { while(1); return 0; }

你琢磨琢磨？这个java程序你准备怎么写。

有个事情，比较好玩儿，如果你真写出这样一个java程序，你就牛逼了。为啥呢，考虑如下程序：

    int main() {
        bigint n,a,b,c;

        for(n = 1;;n++) {
            for(a = 1;a < n;a++) {
                for(b = 1;b < n;b++) {
                    for(c = 1;c < n;c++) {
                        if(a * a * a + b * b * b == c * c * c) {
                            return 0;
                        }
                    }
                }
            }
        }

        // unreachable
    }

你要输出 **YES** ，说明费马大定理不成立，你要输出 **NO** ，说明费马大定理成立，哈哈，你要获得菲尔兹奖了。当然得是前几年，因为这个问题已经被数学家解决了。
